```python
# q19 Implement hypothesis testing using Z-statistics for a sample dataset
import numpy as np
import scipy.stats as stats
data = np.random.normal(0, 1, 1000)
mean = np.mean(data)
std = np.std(data)
z_score = (mean - 0) / (std / np.sqrt(len(data)))
p_value = stats.norm.sf(abs(z_score))
print(p_value)
if p_value < 0.05:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
```

```python
# q20 Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
import scipy.stats as st
data = np.random.normal(0, 1, 1000)
mean = np.mean(data)
std = np.std(data)
n = len(data)
alpha = 0.05
t_critical = st.t.ppf(1 - alpha / 2, n - 1)
margin_of_error = t_critical * std / np.sqrt(n)
confidence_interval = (mean - margin_of_error, mean + margin_of_error)
print(confidence_interval)
```

```python
# q21  Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
import numpy as np
import scipy.stats as st
data = np.random.normal(0, 1, 1000)
mean = np.mean(data)
std = np.std(data)
n = len(data)
alpha = 0.05
t_critical = st.t.ppf(1 - alpha / 2, n - 1)
margin_of_error = t_critical * std / np.sqrt(n)
confidence_interval = (mean - margin_of_error, mean + margin_of_error)
print(confidence_interval)
```

```python
# q22  Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution
import numpy as np
import matplotlib.pyplot as plt
mean = 0
std = 1
x = np.linspace(-4, 4, 100)
y = (1 / (std * np.sqrt(2 * np.pi))) * np.exp(-((x - mean)**2) / (2 * std**2))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('PDF')
plt.title('Normal Distribution PDF')
plt.show()
```
