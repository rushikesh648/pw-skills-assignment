```python
# Q1) . Create a table called employees with the following structure?
#: emp_id (integer, should not be NULL and should be a primary key)Q
#: emp_name (text, should not be NULL)Q
#: age (integer, should have a check constraint to ensure the age is at least 18)Q
#: email (text, should be unique for each employee)Q
#: salary (decimal, with a default value of 30,000).

#Write the SQL query to create the above table with all constraints.

cursor.execute("CREATE DATABASE IF NOT EXISTS employee_db")
cursor.execute("USE employee_db")
cursor.execute("""
    CREATE TABLE IF NOT EXISTS employees (
        emp_id INT NOT NULL PRIMARY KEY,
        emp_name VARCHAR(255) NOT NULL,
        age INT CHECK (age >= 18),
        email VARCHAR(100) UNIQUE,
        salary DECIMAL(10,2) DEFAULT 30000
    )
""")

cursor.close()
conn.close()
```

Q2)  Explain the purpose of constraints and how they help maintain data integrity in a database. Provide
examples of common types of constraints.

ans:

Constraints in a database are rules or restrictions applied to table columns to enforce data integrity and ensure that the data stored is accurate, consistent, and reliable. They are essential in maintaining the quality of the database and preventing invalid or inconsistent data from being entered.

Purpose of Constraints:
Ensuring Data Accuracy: Constraints validate that only valid data is allowed in the database.

Maintaining Consistency: They ensure the data adheres to predefined rules across the database.

Supporting Relationships: In relational databases, constraints maintain relationships between tables, ensuring referential integrity.

Preventing Errors: By restricting certain types of data input, constraints reduce the likelihood of human errors.

Common Types of Constraints:
Here are some widely used constraints with examples:

NOT NULL: Ensures that a column cannot have null (empty) values.

Example: emp_name TEXT NOT NULL ensures every employee has a name.

UNIQUE: Ensures all values in a column are distinct, avoiding duplicates.

Example: email TEXT UNIQUE ensures no two employees share the same email address.

PRIMARY KEY: Combines NOT NULL and UNIQUE to uniquely identify each record in a table.

Example: emp_id INTEGER PRIMARY KEY ensures each employee has a unique ID.

FOREIGN KEY: Links a column in one table to the primary key of another table, maintaining referential integrity.

Example: In a table departments, dept_id may be a foreign key in the employees table.

CHECK: Validates that values in a column satisfy a specific condition.

Example: age INTEGER CHECK (age >= 18) ensures employees are at least 18 years old.

DEFAULT: Assigns a default value to a column when no value is provided.

Example: salary DECIMAL DEFAULT 30000 ensures a default salary of 30,000 for employees.

INDEX (Implicitly Created by Constraints): Often created by constraints like PRIMARY KEY and UNIQUE to improve data retrieval speed.

Q4)  Explain the steps and SQL commands used to add or remove constraints on an existing table. Provide an
example for both adding and removing a constraint.

ans:

Adding and removing constraints on an existing table can be done using ALTER TABLE commands in SQL. Below are the steps and corresponding examples for adding and removing constraints.

1. Adding a Constraint
When adding a new constraint, you use the ALTER TABLE statement along with the ADD CONSTRAINT clause.

Example:
Suppose you have an employees table, and you want to add a CHECK constraint to ensure the age column has values of 18 or more.

 * ALTER TABLE employees
 * ADD CONSTRAINT check_age CHECK (age >= 18);

 2. Removing a Constraint
To remove a constraint, use the ALTER TABLE statement with the DROP CONSTRAINT clause.

Example:

Suppose you want to remove the check_age constraint from the employees table.

 * ALTER TABLE employees
 * DROP CONSTRAINT check_age;

Q5)  Explain the consequences of attempting to insert, update, or delete data in a way that violates constraints.
Provide an example of an error message that might occur when violating a constraint.

ans:

When attempting to insert, update, or delete data in a way that violates constraints in a database, the operation will fail, and an error will be returned by the database management system (DBMS). This safeguards the integrity of the data by ensuring that all predefined rules (constraints) are respected.

Consequences of Violating Constraints:

1. Insert Operation: If you attempt to insert a row that violates any constraints, the database will reject the insertion.

Example: Inserting a row with a NULL value in a NOT NULL column.

2. Update Operation: If you try to update an existing row in a way that breaches constraints, the operation will be aborted.

Example: Updating a column with a value that violates a CHECK condition.

3. Delete Operation: If deleting a row violates a foreign key constraint (such as deleting a parent record that is referenced by child rows), the DBMS will not allow the deletion unless cascading or SET NULL/SET DEFAULT options are configured.

Examples of Errors:
Here are some sample scenarios and potential error messages:

1. Violating NOT NULL Constraint:

>Action: Attempting to insert a record without a required value.

Example:

 * INSERT INTO employees (emp_id, emp_name, age) VALUES (1, NULL, 25);

>Error

 * ERROR:  null value in column "emp_name" violates not-null constraint

2. Violating UNIQUE Constraint:

> Action: Inserting duplicate data in a column with a UNIQUE constraint.

Example:

 * INSERT INTO employees (emp_id, email) VALUES (2, 'example@email.com');
 * INSERT INTO employees (emp_id, email) VALUES (3, 'example@email.com');

>Error:

 * ERROR:  duplicate key value violates unique constraint "employees_email_key"

3. Violating CHECK Constraint:

> Action: Trying to insert a value that doesn't meet the condition in a CHECK constraint.

Example:

 * INSERT INTO employees (emp_id, age) VALUES (4, 16);

>Error:

 * ERROR:  new row for relation "employees" violates check constraint "check_age"

4. Violating FOREIGN KEY Constraint:

> Action: Deleting a parent row that is referenced by a foreign key in another table.

Example:

 * DELETE FROM departments WHERE dept_id = 1;

> Error:

 * ERROR:  update or delete on table "departments" violates foreign key constraint "employees_dept_id_fkey"
on table "employees"
