Q1) What is a Support Vector Machine (SVM).

ans: A Support Vector Machine (SVM) is a powerful supervised machine learning algorithm used for both classification and regression tasks. It's particularly well-suited for complex but small- or medium-sized datasets.

Q2)  What is the difference between Hard Margin and Soft Margin SVM.

ans: Hard Margin SVM:

Assumption: Assumes the data is perfectly separable by a hyperplane (linearly separable) without any errors.
Objective: Aims to find a hyperplane that perfectly classifies all training data points with a maximum margin.
Sensitivity to Outliers: Very sensitive to outliers. Even a single outlier can significantly affect the position of the hyperplane, leading to poor generalization.
Practical Use: Less common in real-world scenarios due to its strict assumption of linear separability.

Soft Margin SVM:

Assumption: Acknowledges that real-world data often has noise and outliers, making perfect separation impossible.
Objective: Aims to find a hyperplane that maximizes the margin while allowing for some misclassifications or violations of the margin.
Sensitivity to Outliers: More robust to outliers compared to hard margin SVMs. It tolerates some misclassifications, making it less affected by noisy data.
Practical Use: More commonly used in practice due to its flexibility and ability to handle real-world datasets with noise and overlapping classes.

Q3) What is the mathematical intuition behind SVM.

ans: w . x + b = 0.

Q4) What is the role of Lagrange Multipliers in SVM.

ans: How Lagrange Multipliers Work in SVMs:

Introducing Lagrange Multipliers: We assign a Lagrange multiplier (αi) to each constraint in the original problem.

Forming the Lagrangian: We construct the Lagrangian function by combining the objective function and the constraints using the Lagrange multipliers:

L(w, b, α) = (1/2) ||w||^2 - Σ αi[yi(w . xi + b) - 1]

Solving for the Dual Problem: We then solve the dual problem of the Lagrangian, which involves finding the optimal values of the Lagrange multipliers (αi). This dual problem is often easier to solve than the original primal problem.

Obtaining the Optimal Hyperplane: Once we have the optimal αi values, we can use them to determine the weight vector (w) and bias term (b) of the optimal hyperplane:

w = Σ αi * yi * xi
   b = yk - w . xk  (for any support vector k)

Q5) What are Support Vectors in SVM.

ans: In an SVM, support vectors are the data points that lie closest to the decision boundary (or hyperplane). They are the most critical elements in determining the optimal hyperplane and play a crucial role in the classification process.

Q6) What is a Support Vector Classifier (SVC).

ans: A Support Vector Classifier (SVC) is a specific type of Support Vector Machine (SVM) algorithm that is used for classification tasks. It's one of the most popular and widely used SVM algorithms.

Q7) What is a Support Vector Regressor (SVR).

ans: A Support Vector Regressor (SVR) is a type of Support Vector Machine (SVM) algorithm that is used for regression tasks, meaning predicting continuous values instead of discrete classes.

Q8) What is the Kernel Trick in SVM.

ans: The Kernel Trick is a crucial technique used in SVMs to handle non-linearly separable data. It allows SVMs to implicitly map data points to a higher-dimensional space where they might become linearly separable, without actually performing the computationally expensive transformation.

Q9) Compare Linear Kernel, Polynomial Kernel, and RBF Kernel.

ans: Linear Kernel:

Formula: K(x, y) = x . y (dot product of input vectors)
Functionality: Creates a linear decision boundary.
Suitable for: Linearly separable data.

Parameters: None

Computational Cost: Low

Pros: Simple, efficient for linearly separable data.

Cons: Not suitable for non-linear data.

Polynomial Kernel:

Formula: K(x, y) = (x . y + c)^d (where 'c' is a constant and 'd' is the degree)

Functionality: Creates a non-linear decision boundary with curves or surfaces.

Suitable for: Non-linearly separable data with moderate complexity.

Parameters: Degree ('d'), constant ('c')

Computational Cost: Moderate, depends on the degree.

Pros: Can handle non-linear data, flexible with degree and constant parameters.

Cons: Can be computationally expensive for high degrees, prone to overfitting if parameters are not chosen carefully.

Radial Basis Function (RBF) Kernel:

Formula: K(x, y) = exp(-||x - y||^2 / (2σ^2)) (where 'σ' is the bandwidth)

Functionality: Creates a non-linear decision boundary with smooth curves or surfaces.

Suitable for: Non-linearly separable data with complex relationships.

Parameters: Bandwidth ('σ')

Computational Cost: Moderate

Pros: Versatile, can handle complex non-linear data, less prone to overfitting compared to polynomial kernel.

Cons: Can be sensitive to the bandwidth parameter, requires careful tuning.

Q10) What is the effect of the C parameter in SVM.

ans: The C parameter in SVMs is a regularization parameter that controls the trade-off between maximizing the margin and minimizing classification errors. It's a crucial hyperparameter that can significantly influence the performance of your SVM model.

Q11) What is the role of the Gamma parameter in RBF Kernel SVM.

ans: The Gamma parameter in an RBF Kernel SVM controls the influence of a single training example on the decision boundary. It defines how far the influence of a single training example reaches.

Q12) What is the Naïve Bayes classifier, and why is it called "Naïve".

ans: The Naïve Bayes classifier is a simple but effective probabilistic machine learning algorithm based on Bayes' Theorem. It's widely used for classification tasks, particularly in text classification and spam filtering.

Why is it called "Naïve"?

The classifier is called "naïve" because of its strong independence assumption among the features. It assumes that the features are conditionally independent, meaning that the presence or absence of one feature does not affect the presence or absence of other features, given the class label.

Q13) What is Bayes’ Theorem.

ans: Bayes' Theorem is a fundamental concept in probability theory and statistics that describes how to update the probability of a hypothesis based on new evidence. It's a powerful tool for making inferences and predictions in situations where we have prior knowledge or beliefs.

Q14)  Explain the differences between Gaussian Naïve Bayes, Multinomial Naïve Bayes, and Bernoulli Naïve Bayes.

ans: Gaussian Naïve Bayes:


Assumption: Assumes that the features follow a Gaussian (normal) distribution.

Suitable for: Continuous data, such as numerical features like height, weight, temperature, etc.

How it works: It calculates the mean and standard deviation of each feature for each class and then uses these parameters to estimate the probability of a data point belonging to a particular class.

Multinomial Naïve Bayes:

Assumption: Assumes that the features represent the frequency of discrete events, such as word counts in text documents.

Suitable for: Discrete data, especially text classification problems where features are word frequencies or TF-IDF vectors.

How it works: It uses the frequency of features in each class to estimate the probability of a data point belonging to a particular class. It often incorporates smoothing techniques (like Laplace smoothing) to handle unseen feature values.

Bernoulli Naïve Bayes:

Assumption: Assumes that the features are binary (present or absent), such as the presence or absence of words in a document.

Suitable for: Binary data or text classification problems where features are represented as binary indicators (e.g., word presence/absence).

How it works: It calculates the probability of each feature being present or absent in each class and then uses these probabilities to estimate the probability of a data point belonging to a particular class.

Q15) When should you use Gaussian Naïve Bayes over other variants.

ans: You should use Gaussian Naïve Bayes when your dataset primarily consists of continuous numerical features that follow a normal (Gaussian) distribution.

Here's a breakdown of situations where Gaussian Naïve Bayes is preferred:

Continuous Features: If your dataset has features like height, weight, temperature, blood pressure, or any other numerical values that can take on a range of values, Gaussian Naïve Bayes is a good choice.

Normal Distribution: Gaussian Naïve Bayes assumes that the features follow a normal distribution. While this assumption might not always be perfectly true in real-world data, it often works well in practice, especially when the features are approximately normally distributed.

Simplicity and Efficiency: Gaussian Naïve Bayes is relatively simple to implement and computationally efficient, making it a good choice for large datasets.

Q16) What are the key assumptions made by Naïve Bayes.

ans: In real-world datasets, features are often correlated or dependent on each other. For example, in a dataset about weather, the features "temperature" and "humidity" are likely to be related. However, Naïve Bayes assumes that these features are independent, which simplifies the calculations but might not reflect the true relationships in the data.

Other Assumptions:

Feature Importance: Naïve Bayes assumes that all features are equally important in contributing to the classification decision.

Data Distribution: Depending on the variant of Naïve Bayes, there are specific assumptions about the

distribution of features:

Gaussian Naïve Bayes: Assumes that continuous features follow a Gaussian (normal) distribution.

Multinomial Naïve Bayes: Assumes that discrete features represent the frequency of events.

Bernoulli Naïve Bayes: Assumes that features are binary (present/absent).

Q17) What are the advantages and disadvantages of Naïve Bayes.

ans: Advantages:

Simplicity and Speed: Naïve Bayes is easy to understand and implement, making it a good choice for quick prototyping or when computational resources are limited. It's also relatively fast, especially for large datasets.

Works well with high-dimensional data: It can handle datasets with many features, which is often the case in text classification and other domains.

Good for text classification and spam filtering: Naïve Bayes has proven to be effective in text classification tasks, such as spam filtering, sentiment analysis, and topic classification.

Handles missing values well: It can handle missing values by simply ignoring them during probability calculations.

Requires less training data: Compared to more complex algorithms, Naïve Bayes can often achieve good performance with relatively less training data.

Performs well with categorical features: It's well-suited for datasets with categorical features, which are common in many real-world applications.

Disadvantages:

Naïve Assumption: The independence assumption of features can be a limitation when features are correlated. This can lead to inaccurate predictions in some cases.

Zero Frequency Problem: If a feature value is not seen in the training data, it assigns a probability of zero, which can affect predictions. This can be addressed using smoothing techniques like Laplace smoothing.

Limited Expressiveness: Due to the independence assumption, Naïve Bayes might not capture complex relationships between features, which can limit its performance in some tasks.

Sensitivity to irrelevant features: Including irrelevant features can negatively impact the accuracy of Naïve Bayes. Feature selection techniques can help mitigate this issue.

Q18) Why is Naïve Bayes a good choice for text classification.

ans: Naïve Bayes is often a preferred choice for text classification tasks due to several reasons:

High Dimensionality: Text data is typically represented as a bag-of-words or TF-IDF vectors, which result in a high-dimensional feature space. Naïve Bayes handles high-dimensional data efficiently and performs well even with a large number of features.

Feature Independence Assumption: While the independence assumption of Naïve Bayes might not hold perfectly true for text data (words in a document are often related), it's a reasonable approximation that often works well in practice. The relative frequencies of words within each class can still provide strong signals for classification, even if the words are not completely independent.

Simplicity and Efficiency: Naïve Bayes is computationally efficient, making it suitable for large text datasets. It's also relatively easy to implement and understand, making it a good choice for quick prototyping and experimentation.

```python
# q21 Write a Python program to train an SVM Classifier on the Iris dataset and evaluate accuracy:

from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

iris = datasets.load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

svm_classifier = SVC(kernel='linear')

svm_classifier.fit(X_train, y_train)

y_pred = svm_classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
# q24 Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision boundary:

import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

svm_classifier = SVC(kernel='poly', degree=3)

svm_classifier.fit(X_train, y_train)

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))
Z = svm_classifier.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('SVM with Polynomial Kernel')
plt.show()
```

```python
# q27 Write a Python program to train an SVM Classifier with different C values and compare the decision boundaries visually.

import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

C_values = [0.1, 1, 10, 100]

fig, axes = plt.subplots(2, 2, figsize=(10, 10))
fig.suptitle('SVM with Different C Values')

for i, C in enumerate(C_values):
    svm_classifier = SVC(kernel='linear', C=C)

    svm_classifier.fit(X_train, y_train)

    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))

    Z = svm_classifier.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    ax = axes[i // 2, i % 2]
    ax.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
    ax.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k')
    ax.set_xlabel('Sepal length')
    ax.set_ylabel('Sepal width')
    ax.set_title(f'C = {C}')

plt.tight_layout()
plt.show()
```

```python
# q29 Write a Python program to apply feature scaling before training an SVM model and compare results with unscaled data.

import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

iris = load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

svm_classifier_unscaled = SVC(kernel='linear')
svm_classifier_unscaled.fit(X_train, y_train)
y_pred_unscaled = svm_classifier_unscaled.predict(X_test)
accuracy_unscaled = accuracy_score(y_test, y_pred_unscaled)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

svm_classifier_scaled = SVC(kernel='linear')
svm_classifier_scaled.fit(X_train_scaled, y_train)
y_pred_scaled = svm_classifier_scaled.predict(X_test_scaled)
accuracy_scaled = accuracy_score(y_test, y_pred_scaled)

print(f"Accuracy without scaling: {accuracy_unscaled}")
print(f"Accuracy with scaling: {accuracy_scaled}")
```

```python
# q30 Write a Python program to train a Gaussian Naïve Bayes model and compare the predictions before and after Laplace Smoothing.

import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

iris = load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

gnb_no_smoothing = GaussianNB(var_smoothing=0)
gnb_no_smoothing.fit(X_train, y_train)
y_pred_no_smoothing = gnb_no_smoothing.predict(X_test)
accuracy_no_smoothing = accuracy_score(y_test, y_pred_no_smoothing)

gnb_smoothing = GaussianNB()
gnb_smoothing.fit(X_train, y_train)
y_pred_smoothing = gnb_smoothing.predict(X_test)
accuracy_smoothing = accuracy_score(y_test, y_pred_smoothing)
print(f"Accuracy without smoothing: {accuracy_no_smoothing}")
print(f"Accuracy with smoothing: {accuracy_smoothing}")

sample_index = 0
print(f"Prediction without smoothing for sample {sample_index}: {y_pred_no_smoothing[sample_index]}")
print(f"Prediction with smoothing for sample {sample_index}: {y_pred_smoothing[sample_index]}")
```

```python
# q31 Write a Python program to train an SVM Classifier and use GridSearchCV to tune the hyperparameters (C,gamma, kernel).

from sklearn import datasets
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

iris = datasets.load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [1, 0.1, 0.01, 0.001],
    'kernel': ['linear', 'rbf', 'poly']
}

svm_classifier = SVC()

grid_search = GridSearchCV(svm_classifier, param_grid, cv=5, scoring='accuracy')

grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
print(f"Best Hyperparameters: {best_params}")

best_estimator = grid_search.best_estimator_

y_pred = best_estimator.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
# q32 Write a Python program to train an SVM Classifier on an imbalanced dataset and apply class weighting and check it improve accuracy.

from sklearn.svm import SVC
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X, y = make_classification(n_samples=1000, n_features=20,
                           n_informative=15, n_redundant=5,
                           n_classes=2, weights=[0.9, 0.1],
                           random_state=42)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

svm_no_weight = SVC(kernel='linear')
svm_no_weight.fit(X_train, y_train)
y_pred_no_weight = svm_no_weight.predict(X_test)
accuracy_no_weight = accuracy_score(y_test, y_pred_no_weight)

svm_weight = SVC(kernel='linear', class_weight='balanced')
svm_weight.fit(X_train, y_train)
y_pred_weight = svm_weight.predict(X_test)
accuracy_weight = accuracy_score(y_test, y_pred_weight)

print(f"Accuracy without class weighting: {accuracy_no_weight}")
print(f"Accuracy with class weighting: {accuracy_weight}")
```

```python
# q35 Write a Python program to perform feature selection before training a Naïve Bayes classifier and compare results.

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target

X_train, X_test, y_train, y_test = train_test_split(df[iris.feature_names], df['target'], test_size=0.3, random_state=42)

selector = SelectKBest(chi2, k=2)
X_train_selected = selector.fit_transform(X_train, y_train)
X_test_selected = selector.transform(X_test)

nb_all_features = GaussianNB()
nb_all_features.fit(X_train, y_train)
y_pred_all_features = nb_all_features.predict(X_test)
accuracy_all_features = accuracy_score(y_test, y_pred_all_features)

nb_selected_features = GaussianNB()
nb_selected_features.fit(X_train_selected, y_train)
y_pred_selected_features = nb_selected_features.predict(X_test_selected)
accuracy_selected_features = accuracy_score(y_test, y_pred_selected_features)
print(f"Accuracy with all features: {accuracy_all_features}")
print(f"Accuracy with selected features: {accuracy_selected_features}")
print("Selected Features:", df.columns[selector.get_support(indices=True)].tolist())
```

```python
# q36 Write a Python program to train an SVM Classifier using One-vs-Rest (OvR) and One-vs-One (OvO) strategies on the Wine dataset and compare their accuracy.

from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.multiclass import OneVsRestClassifier, OneVsOneClassifier
from sklearn.metrics import accuracy_score

wine = datasets.load_wine()
X = wine.data
y = wine.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
ovr_classifier = OneVsRestClassifier(SVC(kernel='linear'))
ovr_classifier.fit(X_train, y_train)
ovr_predictions = ovr_classifier.predict(X_test)
ovr_accuracy = accuracy_score(y_test, ovr_predictions)

ovo_classifier = OneVsOneClassifier(SVC(kernel='linear'))
ovo_classifier.fit(X_train, y_train)
ovo_predictions = ovo_classifier.predict(X_test)
ovo_accuracy = accuracy_score(y_test, ovo_predictions)

print(f"One-vs-Rest (OvR) Accuracy: {ovr_accuracy}")
print(f"One-vs-One (OvO) Accuracy: {ovo_accuracy}")
```

```python
# q39 Write a Python program to train a Naïve Bayes classifier using different prior probabilities and compare performance.

import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

iris = load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

priors = [
    [0.33, 0.33, 0.34],
    [0.5, 0.3, 0.2],
    [0.1, 0.8, 0.1]
]

for prior in priors:
    nb_classifier = GaussianNB(priors=prior)
    nb_classifier.fit(X_train, y_train)
    y_pred = nb_classifier.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Prior probabilities: {prior}")
    print(f"Accuracy: {accuracy}")
    print("-" * 20)
```

```python
# q40 Write a Python program to perform Recursive Feature Elimination (RFE) before training an SVM Classifier and compare accuracy.

from sklearn.svm import SVC
from sklearn.datasets import load_iris
from sklearn.feature_selection import RFE
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

iris = load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

svm_classifier = SVC(kernel='linear')

selector = RFE(estimator=svm_classifier, n_features_to_select=2, step=1)

selector = selector.fit(X_train, y_train)

X_train_selected = selector.transform(X_train)
X_test_selected = selector.transform(X_test)

svm_all_features = SVC(kernel='linear')
svm_all_features.fit(X_train, y_train)
y_pred_all_features = svm_all_features.predict(X_test)
accuracy_all_features = accuracy_score(y_test, y_pred_all_features)

svm_selected_features = SVC(kernel='linear')
svm_selected_features.fit(X_train_selected, y_train)
y_pred_selected_features = svm_selected_features.predict(X_test_selected)
accuracy_selected_features = accuracy_score(y_test, y_pred_selected_features)

print(f"Accuracy with all features: {accuracy_all_features}")
print(f"Accuracy with selected features: {accuracy_selected_features}")
print("Selected Features:", selector.support_)
print("Feature Ranking:", selector.ranking_)
```

```python
# q46  Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import precision_recall_curve, average_precision_score

breast_cancer = load_breast_cancer()
X = breast_cancer.data
y = breast_cancer.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

svm_classifier = SVC(kernel='linear', probability=True)

svm_classifier.fit(X_train, y_train)

y_pred_probs = svm_classifier.predict_proba(X_test)[:, 1]

precision, recall, thresholds = precision_recall_curve(y_test, y_pred_probs)

average_precision = average_precision_score(y_test, y_pred_probs)

plt.figure(figsize=(8, 6))
plt.step(recall, precision, color='b', alpha=0.2, where='post')
plt.fill_between(recall, precision, step='post', alpha=0.2, color='b')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.title(f'Precision-Recall curve: AP={average_precision:0.2f}')
plt.show()
```
