```python
# q10) Write a query to display the order_id, customer_name, and the quantity of products ordered by each customer using an INNER JOIN between all three tables.

import mysql.connector
conn = mysql.connector.connect(user='root', password='root', host='localhost')
cursor = conn.cursor()

cursor.execute("CREATE DATABASE IF NOT EXISTS order_database")
cursor.execute("USE order_database")
cursor.execute('''CREATE TABLE IF NOT EXISTS orders (order_id INT PRIMARY KEY, customer_name VARCHAR(50), product_name VARCHAR(50))''')
cursor.execute('''CREATE TABLE IF NOT EXISTS products (product_id INT PRIMARY KEY, product_name VARCHAR(50), quantity INT)''')
cursor.execute('''CREATE TABLE IF NOT EXISTS order_details (order_id INT PRIMARY KEY, product_id INT, quantity INT)''')

cursor.execute("INSERT INTO orders (order_id, customer_name, product_name) VALUES (146, 'Alise', 'Laptop'), (145, 'Bob', 'Phone')")
cursor.execute("INSERT INTO products (product_id, product_name, quantity) VALUES (101, 'Laptop', 5), (102, 'Phone',10)")
cursor.execute("INSERT INTO order_details (order_id, product_id, quantity) VALUES (106, 101, 2), (108, 102, 3)")

conn.commit()

cursor.close()
conn.close()
```

```python
import mysql.connector
conn = mysql.connector.connect(user='root', password='root', host='localhost', database='order_database')
cursor = conn.cursor()

cursor.execute("SELECT orders.order_id, orders.customer_name, products.quantity FROM orders INNER JOIN order_details ON orders.order_id = order_details.order_id INNER JOIN products ON order_details.product_id = products.product_id")
result = cursor.fetchall()
for row in result:
    print(row)

    cursor.close()
    conn.close()
```

Q2.  Second Normal Form (2NF):

 a. Choose a table in Sakila and describe how you would determine whether it is in 2NF.

 If it violates 2NF, explain the steps to normalize it.

 ans: a. Choosing a Table and Determining 2NF:

I'll choose the film_actor table for this analysis. This table stores the relationship between films and actors.

Determining 2NF:

Check for 1NF: The film_actor table is in 1NF if it has no repeating groups. Since each column contains only atomic values, it likely satisfies 1NF.

Identify Primary Key: The primary key of film_actor is a composite key consisting of film_id and actor_id.

Identify Functional Dependencies: We need to determine the functional dependencies within the table. For

example:

film_id, actor_id -> last_update

This means that the last_update attribute is dependent on both parts of the primary key.

Check for Partial Dependencies: A partial dependency occurs when a non-prime attribute (an attribute not part of the primary key) depends on only a part of the composite key. In this case, last_update depends on both film_id and actor_id (the entire primary key). Therefore, there are no partial dependencies.

Conclusion: Since the film_actor table is in 1NF and has no partial dependencies, it's in 2NF.

If it Violates 2NF (Example with Another Table):

Let's consider a hypothetical table called
rental_details with the following attributes:

rental_id (primary key)
customer_id
film_id
customer_name
film_title
rental_date

Functional Dependencies:

rental_id -> customer_id, film_id, rental_date
customer_id -> customer_name
film_id -> film_title

Violation of 2NF:

We can see that customer_name and film_title are non-prime attributes that depend on parts of the primary key (customer_id and film_id respectively). This creates partial dependencies, violating 2NF.

Steps to Normalize:

Create New Tables:

Create a customers table with customer_id as the primary key and customer_name as an attribute.
Create a films table with film_id as the primary key and film_title as an attribute.

Modify Existing Table:

Remove customer_name and film_title from the rental_details table.
Keep rental_id, customer_id, film_id, and rental_date in the rental_details table.

Establish Foreign Keys:

In the rental_details table, set customer_id as a foreign key referencing the customers table.
In the rental_details table, set film_id as a foreign key referencing the films table.

Q3 Third Normal Form (3NF):

 a. Identify a table in Sakila that violates 3NF. Describe the transitive dependencies

 present and outline the steps to normalize the table to 3NF.

 ans: a. Identifying a Table Violating 3NF

The customer table is a potential candidate for violating 3NF. Let's examine it:

Attributes:

customer_id (primary key)
store_id
first_name
last_name
email
address_id
active
create_date
last_update

Transitive Dependencies:

A transitive dependency occurs when a non-prime attribute depends on another non-prime attribute. In the customer table, we can identify the following:

address_id -> city_id (from the address table)
city_id -> country_id (from the city table)
This means that address_id determines city_id, and city_id further determines country_id. Ultimately, address_id indirectly determines country_id, which is a transitive dependency.

Violation of 3NF:

Since country_id is a non-prime attribute and depends on another non-prime attribute (address_id), the customer table violates 3NF.

Steps to Normalize to 3NF:

Create a New Table: Create a new table called
customer_address with the following attributes:

customer_id (primary key, foreign key referencing customer)
address_id (foreign key referencing address)

Modify the customer Table:

Remove address_id from the customer table.
Keep the remaining attributes in the customer table.

Establish Foreign Keys:

In the customer_address table, set customer_id as a foreign key referencing the customer table.
In the customer_address table, set address_id as a foreign key referencing the address table.
